```python
#imports
from english_words import get_english_words_set
import string
```

```python
#creating 5 letter word array 
web2lowerset = get_english_words_set(['web2'],alpha=True, lower=True)
five_words=[a for a in web2lowerset if len(a) == 5]

known=[]

#creating new information map to track wordle information
info=[]
for i in string.ascii_lowercase:
    info.append([0,1,2,3,4])

#game loop
while True:
    maybes=[]

    #input questions
    word=input("what did you guess: ")
    if word=='': break #game end condition

    green_input=input("what letters are newly green: ")
    green = [x for x in green_input]

    yellow_input=input("what letters are newly yellow: ")
    yellow = [x for x in yellow_input]

    grey_input=input("what letters are newly grey: ")
    grey = [x for x in grey_input]
    
    #making changes to the info map for each green letter
    for letter in green:
        if green==['']:
            break
        for i in range(0,len(info)):
            if string.ascii_lowercase[i]!=letter and word.index(letter) in info[i]:
                info[i].remove(word.index(letter))
            else:
                known.append(letter)
    
    #making changes to the info map for each yellow letter
    for letter in yellow:
        if yellow==['']:
            break
        try:
            info[string.ascii_lowercase.index(letter)].remove(word.index(letter))
        except:
            pass
        if letter not in known or known.count(letter)<yellow.count(letter):
            known.append(letter)
    
    #making changes to the info map for each grey letter
    for letter in grey:
        if grey==['']:
            break
        info[string.ascii_lowercase.index(letter)]=[]

    #gathering possibles words from five_words array using the info map information
    for guess in [i for i in five_words if all([False for letter in known if known.count(letter)!=i.count(letter)])]: #filtering all words tha dont contain all letters that are known
        temp=known.copy()
        for letter in guess:
            if all(i in info[string.ascii_lowercase.index(letter)] for i in [i for i, x in enumerate(guess) if x == letter]): #checking if all instances of a letter from guess are in allowed indexes in info
                pass
            else:
                if guess in five_words:
                    five_words.remove(guess)
        if guess in five_words:
            maybes.append(guess)

    #printing the list of possible words for the next guess
    maybes.sort()
    print(len(maybes),maybes)
    print(known)
```
